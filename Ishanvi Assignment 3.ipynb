### 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

### 11. Build the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
linear_reg = LinearRegression()
decision_tree_reg = DecisionTreeRegressor()
random_forest_reg = RandomForestRegressor()
lasso_reg = Lasso()
ridge_reg = Ridge()
```

### 12. Train the Model

```python
models = [linear_reg, decision_tree_reg, random_forest_reg, lasso_reg, ridge_reg]

def train_models(model):
    model.fit(X_train, y_train)
```

```python
for model in models:
    train_models(model)
```

### 13. Test the Model and 14. Measure the performance using Metrics

```python
def calculate_metrics(model):
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    mae=mean_absolute_error(y_test,y_pred)
    return rmse, r2, mae
```

```python
for model in models:
    rmse, r2, mae= calculate_metrics(model)
    print(type(model).__name__)
    print('RMSE:', rmse)
    print('R2 Score:', r2)
    print('Mean Absolute Error', mae)
    print()
```

```python
rmse_scores = []
r2_scores = []
mae_scores = []
for model in models:
    rmse, r2, mae = calculate_metrics(model)
    rmse_scores.append(rmse)
    r2_scores.append(r2)
    mae_scores.append(mae)

plt.figure(figsize=(10, 5))
plt.title('RMSE Score Comparison')
plt.xlabel('Model')
plt.ylabel('RMSE Score')
plt.bar([type(model).__name__ for model in models], rmse_scores)
plt.show()

plt.figure(figsize=(10, 5))
plt.title('R2 Score Comparison')
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.bar([type(model).__name__ for model in models], r2_scores)
plt.show()

plt.figure(figsize=(10, 5))
plt.title('Mean Absolute Error Comparison')
plt.xlabel('Model')
plt.ylabel('Mean Absolute Error')
plt.bar([type(model).__name__ for model in models], mae_scores)
plt.show()
```
